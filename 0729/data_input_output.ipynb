# 데이터 입출력

## 사용자 입력(input)

> input([prompt])
- 사용자가 입력한 값을 문자열로 반환
- prompt 인자가 있다면, 해당 인자를 출력

````python
user_input = input()
````

````python
user_input = input('너의 이름은?')
````

````python
print(type(user_input))
````

````python
# input()을 통해 입력받은 값은 모두 문자열(string) 형태로 반환

num = input('숫자를 입력 해라! : ')          # 숫자를 입력 해라! : 100
    
print(num)                                  # '100'

print(type(num))                            # <class 'str'>
````

````python
# input()으로 입력 받은 값을 숫자(Number) 타입으로 형 변환
number = int(input('숫자를 입력하라! :'))

print(number)
print(type(number))
````

````python
# 변수 x, y에 각각 정수를 입력 받고 두 값을 더한 값을 출력
x = int(input())
y = int(input())
print(x + y)
````

## split 활용
> .split()
- 문자열을 특정한 단위로 나누어 리스트로 반환
- 기본형 : 'string method'.split()
  - 공백을 기준으로 나누어 리스트로 반환
- 단위 지정 : 'string method_underscore'.split('_')
  - 특정 단위를 기준으로 나누어 리스트로 반환

````python
'apple banana'.split()      # ['apple', 'banana']
````

````python
input('숫자를 입력하라! :').split()     # ['1', '2', '3', '4', '5', '6']
````

## sequence unpacking 응용

- Python에서는 tuple과 list의 요소를 여러 변수에 할당 할 수 있다. 

  ```python
  a, b = ['10', '20']

  print(a, b)     # 10 20
  ```

## map
> map(fuction, iterable, ...)
- iterable의 모든 항목에 지정한 함수를 적용하여 반환

````python
map(int, input('숫자 2개를 입력하라!').split())     # <map at 0x29e75746310>
````

````python
list(map(int, input('숫자 2개를 입력하라!').split()))       # [1, 2, 3]
````

## [실습] 여러번 입력 받기

> 반복문을 사용해 값을 여러번 입력 받아 리스트에 담기

````python
times = int(input('몇 번 입력?'))
data = []
for i in range(times):
    data.append(list( map(int, input('숫자를 입력하라').split()) ))

print(data)
````
